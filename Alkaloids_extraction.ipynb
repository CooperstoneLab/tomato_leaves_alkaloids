# Target Peak Extraction with [ms-mint](https://github.com/LewisResearchGroup/ms-mint)

## Importing libraries

```python
# ms-mint: Instance for MS spectrometry data
from ms_mint.Mint import Mint
import pandas as pd # Data wrangling
#from ms_mint.notebook import Mint
mint = Mint()
```

## Selecting files

In this first example, we are going to use the QC files to exemplify how the targeted extraction is going to be conducted for all the samples.

```python
mint.load_files('./Data/QC_pos*')
mint.ms_files
```

## Import list of targets 

```python
mint.load_targets('targets.csv')
mint.targets['intensity_threshold'] = 0
mint.targets
```

## Run data extraction

```python
mint.run() # run and save in memory
```

## Checking results

```python
mint.results[['ms_file_label', 'peak_label','mz_mean', 'rt', 'peak_area']]
```

```python
mint.plot.peak_shapes()
```

## Saving results table

```python
results_csv = pd.DataFrame(mint.results)
results_csv.to_csv('Results/results_QCs_fixed_rt.csv')
```

## Optimimum retention time

Although the retention time is already optimized fro the given ions, let's see what ms-mint can tells us.

```python
mint.opt.rt_min_max(
    peak_labels=['Dehydrotomatine', 'alpha-Tomatine'],
    plot=True, rel_height=0.7, sigma=50, col_wrap=1, aspect=3,
    height=4
)
```

## Updating optimized retention time

```python
mint.run()
mint.plot.peak_shapes()
```

We see that ms-mint does a good job for dehydrotomatine, but it is cutting off the first scans for tomatine, and introducing a second peak with rt =326, which belongs to a tomatine isomer.

```python
mint.results[['ms_file_label', 'peak_label','mz_mean', 'rt', 'peak_area']]
```
